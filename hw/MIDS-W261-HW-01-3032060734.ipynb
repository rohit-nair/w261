### HW1.0.3.  Bias Variance
What is  bias-variance decomposition in the context machine learning? How is it used in machine learning? 

When trying to predict $Y$ using a trained model $f(X)$, we are interested in determining how well our predictions are. We can do so by calculating expected squared prediction error as:

$$Err(x)=E[(Y−\hat{f} (x))^2]$$

This can be further decomposed into various components as:

$$Err(x) = (E[\hat{f}(x) - f(x))^2 + E[(\hat{f}(x) - E[\hat{f}(x)])^2] + \sigma^2$$

Which can essentially be represented as:

$$Err(x) = Bias^2 + Variance + Irreducible Error$$

The insight this decomposition provides is that any error in our prediction results for these three factors: Bias, Variance and an Irreducible Error. Of these, Bias and Variance are modelling errors while the irreducible error happens due to inherent error in the data for example as caused by error in calliberation of an instrument.

As Bias and Variance are modelling error, we can take steps to reduce their contributions while training a model. Bias occurs when the model is inherently simple and doesn't fit well to the data. This in machine learning is called Underfitting. Variance occurs when the model is comples and has a lots of features that do not address the inherent behavior of the data. In this case too the model doesn't fit the data well and this scenario is called overfitting. Overfitting because while training a model, the model perfoms exceedingly well on the training data as it fits the outliers and noise in the training data while it performs poorly with test data.

This results in a Bias-Variance tradeoff in a model. We can tune our model to either have high bias or high variance while in reality we try to tune the model to where we minimize the error in the model. If we plot Bias and Variance against Model Complexity, we will find that towards the left of the complexity scale, Bias is high and Variance is low while as we move towards right, Variance gets larger as Bias reduces. From the equation above, model error is high lowards the left and the right and the optimum model complexity that minimizes the error is somewhere in between.

```python
%%writefile pWordCount.sh
#!/bin/bash
## pWordCount.sh
## Author: James G. Shanahan
## Usage: pWordCount.sh m wordlist testFile.txt
## Input:
##       m = number of processes (maps), e.g., 4
##       word = a word in quotes, e.g., "alice"
##       inputFile = a text input file
##
## Instructions: Read this script and its comments closely.
##               Do your best to understand the purpose of each command,
##               and focus on how arguments are supplied to mapper.py/reducer.py,
##               as this will determine how the python scripts take input.



usage()
{
    echo ERROR: No arguments supplied
    echo
    echo To run use
    echo "pWordCount.sh m word inputFile"
    echo Input:
    echo "number of processes/maps, EG, 4"
    echo "word = a word in quotes, e.g., 'alice'"
    echo "inputFile = a text input file"
}

if [ $# -eq 0 ]
  then
    usage  
    exit 1
fi
    
## collect user input
m=$1 ## the number of parallel processes (maps) to run

word=$2 ## if set to "*", then all words are used

## a text file 
data=$3

## 'wc' determines the number of lines in the data
## 'perl -pe' regex strips the piped wc output to a number
linesindata=`wc -l $data | perl -pe 's/^.*?(\d+).*?$/$1/'`

## determine the lines per chunk for the desired number of processes
linesinchunk=`echo "$linesindata/$m+1" | bc`

## split the original file into chunks by line
split -l $linesinchunk $data $data.chunk.

## assign python mappers (mapper.py) to the chunks of data
## and emit their output to temporary files
for datachunk in $data.chunk.*; do
    ## feed word list to the python mapper here and redirect STDOUT to a temporary file on disk
    ####
    ####
    ./mapper.py  "$word" < $datachunk | sort -k1,1 > $datachunk.counts &
    ####
    ####
done
## wait for the mappers to finish their work
wait
    
###----------------------------------------------------------------------------------------
#TODO 
#Insert a sort -k1,1 above to collate wordCount records with the same key (i.e., same word)

#
###----------------------------------------------------------------------------------------

    
## 'ls' makes a list of the temporary count files
## 'perl -pe' regex replaces line breaks with spaces
countfiles=`\ls $data.chunk.*.counts | perl -pe 's/\n/ /'`
## feed the list of countfiles to the python reducer and redirect STDOUT to disk
####
####
cat $countfiles | ./reducer.py  > $data.output
####
####

## clean up the data chunks and temporary count files
\rm $data.chunk.*
    
## display the content of the output file:
cat $data.output
```

```python
%%writefile mapper.py
#!/usr/bin/python

# START STUDENT CODE HW12MAPPER
import sys
import re
from collections import defaultdict

count = 0
findword = sys.argv[1]

# If finding count for all the words in the file
if findword == "*":
  wordCounts = defaultdict(int)
  for word in re.findall(r'[a-z]+', sys.stdin.read().lower()):
    wordCounts[word] += 1

  for k in wordCounts:
    print "{}\t{}".format(k, wordCounts[k]) 
#Else get count for particular word being searched for
else:
  count = sys.stdin.read().lower().count(findword.lower())
  print "{}\t{}".format(findword, count) 

# END STUDENT CODE HW12MAPPER
```

```python
%%writefile reducer.py
#!/usr/bin/python

# START STUDENT CODE HW12REDUCER
import sys
import re

current_word = None
count = 0

for line in sys.stdin:
    fragments = re.split(r'\t+', line)
    if len(fragments) == 2:
        c = int(fragments[1])
        if current_word is None:
            current_word = fragments[0]
            count = c
        elif fragments[0] == current_word:
            count += c
        else:
            print "{}\t{}".format(current_word, count)
            current_word = fragments[0]
            count = c
            
print "{}\t{}".format(current_word, count)

# END STUDENT CODE HW12REDUCER
```

```python
%%writefile reducer.py
#!/usr/bin/python

# START STUDENT CODE HW12REDUCER
import sys
import re

lc_count = 0
uc_count = 0

for line in sys.stdin:
    fragments = re.split(r'\t+', line)
    if len(fragments) == 2:
        c = int(fragments[1])
        if fragments[0] == "Lowercase":
            lc_count += c
        else:
            uc_count += c
            
print "{}\t{}".format("Lowercase", lc_count)
print "{}\t{}".format("Uppercase", uc_count)

# END STUDENT CODE HW12REDUCER
```
